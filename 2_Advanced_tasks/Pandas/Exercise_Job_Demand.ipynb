```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_usa = df[df['job_country'] == 'United States'].copy()
df_usa
```

```python
df_usa['job_posted_month'] = pd.to_datetime(df_usa['job_posted_date']).dt.strftime('%B')
df_usa
```

```python
df_usa_pivot =df_usa.pivot_table(index = 'job_posted_month', 
                   columns = 'job_title_short', 
                   aggfunc = 'size')

df_usa_pivot
```

```python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot['month_no'] = pd.to_datetime(df_usa_pivot['job_posted_month'], format='%B').dt.month
df_usa_pivot = df_usa_pivot.sort_values('month_no')
df_usa_pivot = df_usa_pivot.set_index('job_posted_month')
df_usa_pivot = df_usa_pivot.drop(columns='month_no')

df_usa_pivot
```

```python
df_usa_pivot.plot(kind = 'line')
```

```python
top_3 = df_usa['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()

```

```python
df_usa_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the USA')
plt.xlabel('2023')
plt.ylabel('Number of Job Postings')
plt.legend()
plt.show()
```
