```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# Create a copy of the DataFrame and drop rows where the salary_year_avg column has 
# missing values. Display the first 5 rows of the cleaned DataFrame.

df_new = df.copy()
df_new = df_new.dropna(subset = ['salary_year_avg'])
df_new.head(5)
```

```python
# Create a new DataFrame that is a random sample of 10% of the original DataFrame. 
# Display the first 5 rows of the sampled DataFrame.

df_new = df.copy()
df_new = df_new.sample(frac = 0.1)

df_new.head(5)
```

```python
# Create a copy of the original DataFrame and fill in missing values in the salary_year_avg 
# column with the median salary. Display the first 5 rows of the altered DataFrame.

df_copy = df.copy()
median_salary = df_copy['salary_year_avg'].median()
df_copy['salary_year_avg'] = df_copy['salary_year_avg'].fillna(median_salary)

df_copy.head(5)
```
